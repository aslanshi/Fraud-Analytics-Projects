```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Cleaned_data.csv')
```

```python
df = df[['RECORD', 'B','BBLE', 'ZIP', 'BLOCK', 'BLDGCL', 'TAXCLASS', 'STORIES',
            'LTFRONT', 'LTDEPTH', 'BLDFRONT', 'BLDDEPTH', 'FULLVAL', 'AVTOT', 'AVLAND']]
```

```python
null_zero = pd.DataFrame(dict(zip(['# of null', '# of zeros'], [df.isna().sum(), (df == 0).sum()])))
null_zero
```

```python
mean = df.groupby(['ZIP', 'BLDGCL'])['FULLVAL'].mean()
cou = df.groupby(['ZIP', 'BLDGCL'])['FULLVAL'].count()

mean_cou = pd.DataFrame(dict(zip(['fullval mean', 'group size'], [mean, cou])))
mean_cou['null group'] = mean_cou['fullval mean'].apply(lambda x: 'Yes' if pd.isnull(x) else 'No')

null_mean_cou = mean_cou[mean_cou['null group'] == 'Yes']
null_mean_cou
```

```python
mean_cou[mean_cou['group size'] < 5]
```
