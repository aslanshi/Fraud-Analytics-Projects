## Filter

### KS score

```python
ks_score = []
for column in df1:
    ks_score.append(sps.ks_2samp(goods[column], bads[column])[0])
```

### FDR

```python
numbads = len(bads)
topRows = int(round(len(df1)*0.03))
```

```python
df1['check_label'] = df1.label
fdr = []
for column in df1.columns[:-1]:
    temp1 = df1[[column, 'check_label']].sort_values(column, ascending = False).head(topRows)
    temp2 = df1[[column, 'check_label']].sort_values(column, ascending = True).head(topRows)
    FDR1 = temp1.check_label.sum()/numbads
    FDR2 = temp2.check_label.sum()/numbads
    fdr.append(max(FDR1, FDR2))
```

```python
df1.drop('check_label', axis = 1, inplace = True)
```

### Ranking

```python
rank = pd.DataFrame({
    'Field': df1.columns,
    'KS': ks_score,
    'FDR': fdr
})
```

```python
rank['KS rank'] = rank.KS.rank()
rank['FDR rank'] = rank.FDR.rank()
rank['Average rank'] = (rank['KS rank'] + rank['FDR rank'])/2
```

```python
rank.head()
```

```python
rank1 = rank.sort_values('KS rank', ascending = False)
rank1
```

```python
rank2 = rank.sort_values('FDR rank', ascending = False)
rank2
```

```python
rank3 = rank.sort_values('Average rank', ascending = False)
rank3
```

### Stepwise

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
import timeit
```

```python
start = timeit.default_timer()
model = LogisticRegression()
rfecv = RFECV(estimator = model, step = 1, cv = 2, n_jobs = -1, verbose = 3, scoring = 'roc_auc')
rfecv.fit(df2.values, df1.label.values)
print('Total run time: {:.2f}mins'.format((timeit.default_timer() - start)/60))
```
