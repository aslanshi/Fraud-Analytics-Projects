## Filter

### KS score

```python
# sps.ks_2samp(goods.address_count_30, bads.address_count_30)[0]
```

```python
ks_score = []
for column in df1:
    ks_score.append(sps.ks_2samp(goods[column], bads[column])[0])
```

```python
# ks_score
```

### FDR

```python
numbads = len(bads)
topRows = int(round(len(df1)*0.03))
```

```python
## it's much faster if we select only important fields !!!

df1['check_label'] = df1.label
fdr = []
for column in df1.columns[:-1]:
    temp1 = df1[[column, 'check_label']].sort_values(column, ascending = False).head(topRows)
    temp2 = df1[[column, 'check_label']].sort_values(column, ascending = True).head(topRows)
    FDR1 = temp1.check_label.sum()/numbads
    FDR2 = temp2.check_label.sum()/numbads
    fdr.append(max(FDR1, FDR2))
    # print(column + ' Done')
```

```python
df1.drop('check_label', axis = 1, inplace = True)
```

### Ranking

```python
rank = pd.DataFrame({
    'Field': df1.columns,
    'KS': ks_score,
    'FDR': fdr
})
```

```python
rank['KS rank'] = rank.KS.rank()
rank['FDR rank'] = rank.FDR.rank()
rank['Average rank'] = (rank['KS rank'] + rank['FDR rank'])/2
```

```python
rank.head()
```
